# The Monty Hall Problem: Strategies
### This is a statistical analysis of different strategies (switching doors vs keeping initial choice).

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import gridspec
```

### Read in csv files
### These files each contain the results of 10000 simulations of the Monty Hall Problem.
### The 'switch' file contains the results of switching doors every time.
### The 'stay' file contains the results of keeping initial door choice every time.

```python
switch = pd.read_csv('C:\\Users\\mitch\\Downloads\\montyhallswitch.csv')
stay = pd.read_csv('C:\\Users\\mitch\\Downloads\\montyhallstay.csv')
```

```python
switch.head()
```

### The Door Number Column indicates which door the car was behind.

```python
stay.head()
```

### Create a binary outcome column.

```python
switch['Result'] = [0 if outcome == 'goat' else 1 for outcome in switch['Outcome']]
```

```python
switch.head()
```

```python
stay['Result'] = [0 if outcome == 'goat' else 1 for outcome in stay['Outcome']]
```

```python
stay.head()
```

### Create columns to keep running totals of cars won and goats won.

```python
car_count = []
count = 0
for result in switch['Result']:
    if result == 1:
        count += 1
        car_count.append(count)
    else:
        car_count.append(count)

switch['Car Count'] = car_count

goat_count = []
gcount = 0
for result in switch['Result']:
    if result == 0:
        gcount += 1
        goat_count.append(gcount)
    else:
        goat_count.append(gcount)

switch['Goat Count'] = goat_count
```

```python
switch.head(10)
```

```python
car_count_stay = []
count_stay = 0
for result in stay['Result']:
    if result == 1:
        count_stay += 1
        car_count_stay.append(count_stay)
    else:
        car_count_stay.append(count_stay)

stay['Car Count'] = car_count_stay

goat_count_stay = []
gcount_stay = 0
for result in stay['Result']:
    if result == 0:
        gcount_stay += 1
        goat_count_stay.append(gcount_stay)
    else:
        goat_count_stay.append(gcount_stay)

stay['Goat Count'] = goat_count_stay
```

```python
stay.head(10)
```

```python
switch['Unnamed: 0'] = [i + 1 for i in switch['Unnamed: 0']]
```

```python
stay['Unnamed: 0'] = [i + 1 for i in stay['Unnamed: 0']]
```

```python
switch = switch.rename(columns={'Unnamed: 0': 'Play Count'})
stay = stay.rename(columns={'Unnamed: 0': 'Play Count'})
```

### Create columns to keep track of the running totals of cars won and goats won, in percentage terms.

```python
plays = 1
car_pct = []
for car in switch['Car Count']:
    pct = (car/plays) * 100
    pct = np.round(pct, 2)
    car_pct.append(pct)
    plays += 1

switch['Car %'] = car_pct

gplays = 1
goat_pct = []
for goat in switch['Goat Count']:
    pct = (goat/gplays) * 100
    pct = np.round(pct, 2)
    goat_pct.append(pct)
    gplays += 1

switch['Goat %'] = goat_pct
```

```python
switch.head()
```

```python
plays_stay = 1
car_pct_stay = []
for car in stay['Car Count']:
    pct = (car/plays_stay) * 100
    pct = np.round(pct, 2)
    car_pct_stay.append(pct)
    plays_stay += 1

stay['Car %'] = car_pct_stay

gplays_stay = 1
goat_pct_stay = []
for goat in stay['Goat Count']:
    pct = (goat/gplays_stay) * 100
    pct = np.round(pct, 2)
    goat_pct_stay.append(pct)
    gplays_stay += 1

stay['Goat %'] = goat_pct_stay
```

```python
stay.head()
```

### Plot the results.
### The first plot shows the percentage of cars won by switching doors (10000 simulations).
### The second plot shows the percentage of cars won by keeping the initial door choice (10000 simulations).
### The bottom plot shows the percentage of cars won, comparing both strategies for the first 400 simulations.
### The asymptotes in each of the plots represent the expected probability of winning a car using each of the strategies.

```python
asym = [66.6667 for play in switch['Play Count']]
asym2 = [33.3333 for play in stay['Play Count']]

gs = gridspec.GridSpec(2,2)
fig = plt.figure(figsize=(10,10))
fig.suptitle('Strategy Differentiation in Monty Hall')

ax = plt.subplot(gs[0,0])
plt.plot(switch['Play Count'], switch['Car %'], color='red')
plt.plot(switch['Play Count'], asym, '--', color='black')
ax.set_yticks([0, 20, 40, 60, 80, 100])
plt.title('Switching Doors')
plt.ylabel('Percentage of Cars Won')

ax2 = plt.subplot(gs[0,1])
plt.plot(stay['Play Count'], stay['Car %'], color='blue')
plt.plot(stay['Play Count'], asym2, '--', color='black')
ax2.set_yticks([0, 20, 40, 60, 80 , 100])
plt.title('Keeping Door')

ax3 = plt.subplot(gs[1, :])
plt.plot(switch['Play Count'], switch['Car %'], color='red')
plt.plot(switch['Play Count'], stay['Car %'], color='blue')
plt.plot(switch['Play Count'], asym, '--', color='black')
plt.plot(switch['Play Count'], asym2, '--', color='black')
plt.axis([-5, 400, -5, 100])
plt.xlabel('# of Games Simulated')
plt.ylabel('Percentage of Cars Won')
ax3.set_yticks([0, 20, 33.33, 40, 60, 66.67, 80, 100])

plt.savefig('C:\\Users\\mitch\\Documents\\Figures\\montyhall.jpg')
```
